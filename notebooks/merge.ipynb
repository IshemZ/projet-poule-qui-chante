```python
import pandas as pd
```

```python
indices_commerciaux = pd.read_csv("../export/indices_commerciaux_2020.csv", sep=",")
population = pd.read_csv("../export/population_2020.csv", sep=",")
securite_alimentaire = pd.read_csv("../export/securite_alimentaire_2020.csv", sep=",")
groupe_pays = pd.read_csv("../export/groupe_pays.csv", sep=",")

indices_commerciaux
```

```python
# Récupérer les sets de country_code
codes_pop = set(population['country_code'])
codes_sec = set(securite_alimentaire['country_code'])
codes_indices = set(indices_commerciaux['country_code'])

# country_code présents dans population mais pas dans securite_alimentaire
manquants_dans_secu = codes_pop - codes_sec
print("Présents dans population mais pas dans securite_alimentaire :", manquants_dans_secu)

# country_code présents dans securite_alimentaire mais pas dans population
manquants_dans_pop = codes_sec - codes_pop
print("Présents dans securite_alimentaire mais pas dans population :", manquants_dans_pop)

# country_code présents dans indices_commerciaux mais pas dans population
manquants_dans_pop_indices = codes_pop - codes_indices
print("Présents dans population mais pas dans indices_commerciaux :", manquants_dans_pop_indices)

# country_code présents dans population mais pas dans indices_commerciaux
manquants_dans_indices_pop = codes_indices - codes_pop
print("Présents dans indices_commerciaux mais pas dans population :", manquants_dans_indices_pop)
```

```python
# Codes présents uniquement dans population
uniques_pop = codes_pop - (codes_sec | codes_indices)

# Afficher les country_code et zone correspondants
print("country_code et zone présents uniquement dans population :")
print(population[population['country_code'].isin(uniques_pop)][['country_code', 'zone']])
```

```python
df_merged = pd.merge(groupe_pays, population, on='country_code', how='left', suffixes=('_pop', '_secu'))
df_merged = pd.merge(df_merged, securite_alimentaire, on='country_code', how='left', suffixes=('', '_secu'))
df_merged = pd.merge(df_merged, indices_commerciaux, on='country_code', how='left', suffixes=('', '_indices'))

df_merged
```
