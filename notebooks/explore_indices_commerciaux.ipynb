```python
import pandas as pd
from unidecode import unidecode
```

```python
indices_commerciaux = pd.read_csv('../database/indices_commerciaux_2020.csv', dtype=str)
def preprocess_indices_commerciaux_data(indices_commerciaux):
    indices_commerciaux = indices_commerciaux.drop(columns=['Description du Symbole', 'Symbole', 'Code année', 'Produit', 'Code Élément', 'Année', 'Code Produit (FAO)', 'Domaine', 'Code Domaine', 'Unité'])
    return indices_commerciaux
indices_commerciaux = preprocess_indices_commerciaux_data(indices_commerciaux)
indices_commerciaux
```

```python
def convert_indices_commerciaux_to_numeric(indices_commerciaux):
    indices_commerciaux['Valeur'] = indices_commerciaux['Valeur'].astype(float)
    return indices_commerciaux

indices_commerciaux_transformed = convert_indices_commerciaux_to_numeric(indices_commerciaux)
indices_commerciaux_transformed
```

```python
def transform_indices_commerciaux_data(indices_commerciaux):
    grouper_lignes = indices_commerciaux.groupby(['Code zone (FAO)', 'Zone', 'Élément'], as_index=False).sum()
    return grouper_lignes
grouper_lignes = transform_indices_commerciaux_data(indices_commerciaux)
grouper_lignes
```

```python
def transform_indices_commerciaux_data(grouper_lignes):
    indices_commerciaux = grouper_lignes.pivot(index=['Code zone (FAO)','Zone'], columns='Élément', values='Valeur')
    indices_commerciaux = indices_commerciaux.reset_index()
    indices_commerciaux = indices_commerciaux.rename(columns={'Code zone (FAO)': 'country_code'})
    return indices_commerciaux

indices_commerciaux_transformed = transform_indices_commerciaux_data(grouper_lignes)
indices_commerciaux_transformed
```

```python
def transform_regex_indices_commerciaux(indices_commerciaux_transformed):
    indices_commerciaux_transformed.columns = indices_commerciaux_transformed.columns.str.replace(' ', '_', regex=False)
    indices_commerciaux_transformed.columns = indices_commerciaux_transformed.columns.str.lower()
    indices_commerciaux_transformed.columns = [unidecode(col) for col in indices_commerciaux_transformed.columns]
    return indices_commerciaux_transformed

indices_commerciaux_transformed = transform_regex_indices_commerciaux(indices_commerciaux_transformed)
```

```python
def save_transformed_indices_commerciaux_data(indices_commerciaux_transformed):
    indices_commerciaux_transformed.to_csv('../export/indices_commerciaux_2020.csv', index=False, encoding='utf-8')
    print("Indices commerciaux data transformed and saved successfully.")

save_transformed_indices_commerciaux_data(indices_commerciaux_transformed)
```
