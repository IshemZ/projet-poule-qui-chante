```python
import pandas as pd
from unidecode import unidecode
```

```python
groupe_pays = pd.read_csv("../database/groupe_pays_fao.csv", sep=",", dtype=str)
groupe_pays
```

```python
def transform_groupe_pays(groupe_pays):
    groupe_pays.columns = groupe_pays.columns.str.replace(' ', '_', regex=False)
    groupe_pays.columns = groupe_pays.columns.str.lower()
    groupe_pays.columns = [unidecode(col) for col in groupe_pays.columns]
    return groupe_pays

groupe_pays = transform_groupe_pays(groupe_pays)
groupe_pays
```

```python
def save_groupe_pays_csv(groupe_pays):
    groupe_pays.to_csv("../export/groupe_pays.csv", index=False, sep=",", encoding="utf-8-sig")
    print("Groupe Pays data transformed and saved successfully.")

save_groupe_pays_csv(groupe_pays)
```
